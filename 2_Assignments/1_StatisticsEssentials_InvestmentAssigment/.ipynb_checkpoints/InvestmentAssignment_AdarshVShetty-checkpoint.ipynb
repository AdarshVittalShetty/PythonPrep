## Checkpoint 1: Data Cleaning 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Reading companies data to 'companies' dataframe
companies = pd.read_csv('companies.txt',encoding='ISO-8859-1',sep='\t')

# Reading rounds2 data to 'rounds2' dataframe
rounds2 = pd.read_csv('rounds2.csv',encoding='ISO-8859-1')

#How many unique companies are present in rounds2?
#Preparing the data by converting to lower(there might be errors with case sensitivity)
rounds2.company_permalink = rounds2.company_permalink.str.lower()
#stripping leading and trailing white spaces, if any
rounds2.company_permalink = rounds2.company_permalink.str.strip()
rounds2.company_permalink = rounds2.company_permalink.str.encode('ISO-8859-1').str.decode('ascii','ignore')
companies_count_rounds2 = len(rounds2['company_permalink'].unique())
print("Unique companies count in rounds2 table = ",companies_count_rounds2)

#How many unique companies are present in the companies file?
#Preparing the data by converting to lower(there might be errors with case sensitivity)
companies.permalink = companies.permalink.str.lower()
#stripping leading and trailing white spaces, if any
companies.permalink = companies.permalink.str.strip()
companies.permalink = companies.permalink.str.encode('ISO-8859-1').str.decode('ascii','ignore')
companies_count_companies = len(companies['permalink'].unique())
print("Unique companies count in Companies table = ",companies_count_companies)

#In the companies data frame, which column can be used as the Â unique key for each company? Write the name of the column.
print("Total number of rows in companies table = ",companies.permalink.count())

```

## Since, the number of rows and the count of unique company ID(Permalink)  in companies table is same. It means that the each row has a unique permalink value.Hence, 'permalink' can be used as Unique key for each company

```python
s1 = set(rounds2['company_permalink'].unique())
s2 = set(companies['permalink'].unique())
s3 = s1.difference(s2)
print(s3)
```

## since the difference between s1 and s2 is empty, there are no companies in rounds2 which are not present in companies

# Cleaning 'companies' table

```python
# Cleaning Companies dataframe

companies['name'] = companies['name'].str.lower()
companies['name'] = companies['name'].str.strip()
companies.name = companies.name.str.encode('ISO-8859-1').str.decode('ascii','ignore')

# Check is,rows having all missing value
print(companies.isnull().all(axis=1).sum())
print(companies.shape)
print(companies.info())
```

### none of the rows are completely empty

```python
# bringing uniformity to category_list and country_code
#Preparing the data by converting to lower or upper(there might be errors with case sensitivity)
companies.category_list = companies.category_list.str.lower()
companies.country_code = companies.country_code.str.upper()
#stripping leading or trailing white spaces
companies.category_list = companies.category_list.str.strip()
companies.country_code = companies.country_code.str.strip()

# summing up the missing values (column-wise)
round(100*(companies.isnull().sum()/len(companies.index)), 2)
```

```python
# we can drop state_code,region ,city,founded_at and homepage_url .As it is not value added for our Analysis
companies = companies.drop('state_code', axis=1)
companies = companies.drop('region', axis=1)
companies = companies.drop('city', axis=1)
companies = companies.drop('founded_at', axis=1)
companies = companies.drop('homepage_url', axis=1)
round(100*(companies.isnull().sum()/len(companies.index)), 2)
```

```python
print(companies.count())
```

# Cleaning "rounds2" table

```python
print(rounds2.shape)
print(rounds2.info())
round(100*(rounds2.isnull().sum()/len(rounds2.index)), 2)
```

```python
# 72.91% of funding_round_code is missing, let us drop this column
rounds2 = rounds2.drop('funding_round_code', axis=1)
round(100*(rounds2.isnull().sum()/len(rounds2.index)), 2)
```

```python
#Let us check if there are any duplicate entries in the rounds2 table by checking the uniqueness of 'funding_round_permalink'
rounds2.funding_round_permalink = rounds2.funding_round_permalink.str.encode('ISO-8859-1').str.decode('ascii','ignore')
rounds2.funding_round_permalink = rounds2.funding_round_permalink.str.lower()
rounds2.funding_round_permalink = rounds2.funding_round_permalink.str.strip()
rounds2.funding_round_permalink.describe()
```

```python
# Looking at the above result, we can infer that there are no duplicate entries as the count is equal to unique values count
# Hence, we do not have the problem of cleaning duplicate values
```

## Now we have a clean rounds2 table and clean companies table
## Let us merge them

```python
# since the rounds2 table has the funding details,keep the rounds2 table to the left as we do not want to lose data there,for now
master_frame = pd.merge(left=rounds2,right=companies, how='left', left_on='company_permalink', right_on='permalink')
print(master_frame.shape)
print(round(100*(master_frame.isnull().sum()/len(master_frame.index)),2))
```

```python
# raised_amount_usd is missing for few companies.Let us clean the rows which do not have the value.
#Let us see the spread of the data
master_frame['raised_amount_usd'].describe()
```

```python
# Since the spread is too large, and we have lot of data. Loosing 17.39% (rows where raised_amount_US is null)of the rows 
# wouldn't affect the outcome by much.So, let us delete these rows.
# removing NaN raised_amount_US rows
master_frame = master_frame[~pd.isnull(master_frame['raised_amount_usd'])]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# Now, Let us delete the rows where the country_code is missing.
# removing NaN Country_code rows
master_frame = master_frame[~pd.isnull(master_frame['country_code'])]

round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#print(companies.category_list.mode())
master_frame.category_list.describe()
```

```python
# The mode for category_list = biotechnology with freq = 7621, which is < 10% of the total 88529.
#Instead of imputing the data and  biasing the table towards 'biotechnology' Category, let us delete the rows with null category_list
master_frame = master_frame[~pd.isnull(master_frame['category_list'])]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame.shape
```

## Checkpoint 2: Funding Type Analysis

## We can find the average values of funding amount based on investment type  by first grouping them by 'funding_round_type' and then taking the mean of the new dataframe.

```python
#funding_round_type need to be cleaned to do further analysis on investment type

#Preparing the data by converting to lower(there might be errors with case sensitivity)
master_frame.funding_round_type = master_frame.funding_round_type.str.lower()
#stripping leading or trailing white spaces
master_frame.funding_round_type = master_frame.funding_round_type.str.strip()

df_by_funding_round_type = master_frame.groupby('funding_round_type')
print(df_by_funding_round_type['raised_amount_usd'].mean())
```

## Checkpoint 4: Sector Analysis 1

```python
master_frame.shape
master_frame.info()
```

```python
# Extracting the primary sector from the category list into a new column 'primary sector'

master_frame.loc[:,'primary_sector'] = master_frame['category_list'].str.split('|').str[0].str.lower()
#stripping leading or trailing white spaces
master_frame.loc[:,'primary_sector'] = master_frame.primary_sector.str.strip()
master_frame
```

## Cleaning mapping table

```python
mapping = pd.read_csv("mapping.csv")
mapping.head()
```

```python
# drop 'Blanks' column and then delete the rows which has no mapping
mapping = mapping.drop('Blanks', axis=1)
mapping.head()
```

```python
# Now, Let us delete the rows where the category_list is missing.
# removing NaN category_list rows
mapping = mapping[~pd.isnull(mapping['category_list'])]
mapping.head()
```

```python
def map_mainSector(row):
    for c in mapping.columns:
        if row[c]==1:
            return c
        

mapping['main_sector'] = mapping.apply(map_mainSector, axis=1)
mapping_simplified = mapping
mapping_simplified = mapping_simplified.drop('Automotive & Sports', axis=1)
mapping_simplified = mapping_simplified.drop('Cleantech / Semiconductors', axis=1)
mapping_simplified = mapping_simplified.drop('Entertainment', axis=1)
mapping_simplified = mapping_simplified.drop('Health', axis=1)
mapping_simplified = mapping_simplified.drop('Manufacturing', axis=1)
mapping_simplified = mapping_simplified.drop('News, Search and Messaging', axis=1)
mapping_simplified = mapping_simplified.drop('Others', axis=1)
mapping_simplified = mapping_simplified.drop('Social, Finance, Analytics, Advertising', axis=1)
mapping_simplified.category_list = mapping_simplified.category_list.str.lower()
mapping_simplified.loc[:,'category_list'] = mapping_simplified.category_list.str.strip()

mapping_simplified.head()
```

```python
# By checking the mapping.csv file, we can notice 'na' is replace by '0', let us replace 0 with na
# After doing the above step Enterprise 2.0 was changed to Enterprise 2.na, change that back to Enterprise 2.0
mapping_simplified.category_list = mapping_simplified.category_list.replace({'0':'na','2.na':'2.0'},regex=True)
mapping_simplified.head()
```

```python
# Now, let us do a left join of master_frame and mapping_simplified on primary sector and category list
final_master_frame = pd.merge(left=master_frame,right=mapping_simplified, how='left', left_on='primary_sector', right_on='category_list')
final_master_frame.head()
```

```python
#deleting redundant columns
final_master_frame = final_master_frame.drop('category_list_y', axis = 1)
final_master_frame = final_master_frame.drop('permalink', axis = 1)
final_master_frame.head()
```

```python
final_master_frame.info()
```

```python
print(round(100*(final_master_frame.isnull().sum()/len(final_master_frame.index)),2))
```

```python
#since the last set of Analysis revolves around sectors.
#Let us delete the rows which have NaN main_sector
final_master_frame = final_master_frame[~pd.isnull(final_master_frame['main_sector'])]
print(round(100*(final_master_frame.isnull().sum()/len(final_master_frame.index)),2))
```

## Checkpoint 5: Sector Analysis 2 

```python
# D1,D2,D3 are the dataframes created for the top three english speaking nations respectively

# Filter final_master_frame for FT=venture and  5 million USD <= funding_amt <= 15Million USD
final_master_frame = final_master_frame.loc[(final_master_frame.funding_round_type == 'venture') & (final_master_frame.raised_amount_usd >= 5000000) & (final_master_frame.raised_amount_usd <= 15000000)]

# Now separate the above dataframe based on top 3 english speaking countries
D1 = final_master_frame.loc[final_master_frame.country_code == 'USA',:]
D2 = final_master_frame.loc[final_master_frame.country_code == 'GBR',:]
D3 = final_master_frame.loc[final_master_frame.country_code == 'IND',:]


#The total number (or count) of investments for each main sector in a separate column
D1.loc[:,'count'] = D1.groupby('main_sector')['main_sector'].transform('size')
D2.loc[:,'count'] = D2.groupby('main_sector')['main_sector'].transform('size')
D3.loc[:,'count'] = D3.groupby('main_sector')['main_sector'].transform('size')


#The total amount invested in each main sector in a separate column
D1.loc[:,'Total_Amount_main_sector'] = D1.groupby('main_sector')['raised_amount_usd'].transform('sum')
D2.loc[:,'Total_Amount_main_sector'] = D2.groupby('main_sector')['raised_amount_usd'].transform('sum')
D3.loc[:,'Total_Amount_main_sector'] = D3.groupby('main_sector')['raised_amount_usd'].transform('sum')

print("Total number of Investments (count) for FT = 'venture' and funding amount between 5 million to 15 million USD")
print("C1 : " , len(D1.index))
print("C2 : " , len(D2.index))
print("C3 : " , len(D3.index))
```

### Plot3: No. of Investments for top 3 sectors within the country group of Top 3 countries

```python
# 'D1_FT_Venture_Top_Sectors' is the Dtaframe for the chosen Funding_type == venture and top3 sectors
D1_FT_Venture_Top_Sectors = D1.loc[((D1['main_sector'] == 'Others') |  (D1['main_sector'] == 'Social, Finance, Analytics, Advertising') | (D1['main_sector'] == 'Cleantech / Semiconductors'))]
# 'D2_FT_Venture_Top_Sectors' is the Dtaframe for the chosen Funding_type == venture  and top3 sectors
D2_FT_Venture_Top_Sectors = D2.loc[((D2['main_sector'] == 'Others') |  (D2['main_sector'] == 'Social, Finance, Analytics, Advertising') | (D2['main_sector'] == 'Cleantech / Semiconductors'))]
# 'D3_FT_Venture_Top_Sectors' is the Dtaframe for the chosen Funding_type == venture  and top3 sectors
D3_FT_Venture_Top_Sectors = D3.loc[((D3['main_sector'] == 'Others') |  (D3['main_sector'] == 'Social, Finance, Analytics, Advertising') | (D3['main_sector'] == 'News, Search and Messaging'))]

# concatenate the the three filtered and clean data frames
df = pd.concat([D1_FT_Venture_Top_Sectors, D2_FT_Venture_Top_Sectors,D3_FT_Venture_Top_Sectors], axis = 0)

# set figure size for larger figure
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')

# specify hue="main_sector"
sns.barplot(x='country_code', y='count', hue="main_sector", data=df)
#plt.yscale('log')
plt.show()
```
