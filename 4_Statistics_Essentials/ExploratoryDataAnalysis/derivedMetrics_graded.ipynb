```python
import numpy as np
import pandas as pd
import datetime
from datetime import date
df = pd.read_csv("grades.csv")
df.submit_time = pd.to_datetime(df.submit_time)
df['day'] = df.submit_time.dt.day
df['month'] = df.submit_time.dt.month
df['year'] = df.submit_time.dt.year
df['hour'] = df.submit_time.dt.hour
df['min'] = df.submit_time.dt.minute
df.submission = df.submission.str.strip().str.lower()
df['isZip'] = df.submission.apply(lambda x : 0 if x.find('.zip') == -1 else 1)
sum = round(100*df.isZip.sum()/len(df.index),2)
df.head(20)
```

```python
#Jan 3, 2017 - 11:59:59 PM
# datetime(year, month, day, hour, minute, second, microsecond)
first_cutoff = datetime.datetime(2017, 1, 3, 23, 59, 59, 0)
sec_cutoff = datetime.datetime(2017, 1, 9, 23, 59, 59, 0)
df['submittedAfterOne'] = df.submit_time.apply(lambda x : 1 if x>first_cutoff else 0)
df['submittedAfterTwo'] = df.submit_time.apply(lambda x : 1 if x>sec_cutoff else 0)
df.submittedAfterOne.sum()
```
