```python
import numpy as np
import pandas as pd
import datetime

df = pd.read_csv('odi-batting.csv')
df.head()
```

```python
df['century'] = df['Runs'].apply(lambda x: 1 if x >= 100 else 0)
#df1 = df[(df.Player == 'Sachin R Tendulkar') | (df.Player == 'Ricky T Ponting')]
#df.Player = df.Player.str.strip()
#df.Player = df.Player.str.lower()
print(df[df.Player == 'Sachin R Tendulkar']['century'].sum())
print(df[df.Player == 'Ricky T Ponting']['century'].sum())
```

```python
df['SR'] = 100*df['Runs']/df['Balls']
df1 = df[df.century == 1]
df1 = df1.sort_values(['SR'], ascending=False)
df1.head()
```

```python
df['MatchDate']= pd.to_datetime(df['MatchDate'],infer_datetime_format=True)
#df.info()
df['Year']= df.MatchDate.dt.year
df.head()
```
