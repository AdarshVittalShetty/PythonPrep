```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('tendulkar_ODI.csv')
df = df[~(df['Runs'] == 'DNB')]
df = df[~(df['Runs'] == 'TDNB')]

df.Runs=df.Runs.str.strip()
df.Runs=df.Runs.str.replace('*','')
df.Runs = pd.to_numeric(df.Runs)
sns.distplot(df['Runs'])
plt.show()
```
